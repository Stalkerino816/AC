1.2 Load IRIS dataset

```python
#To do: type your code here:
#1.1 imports
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#1.2Loading the iris data
data = load_iris()
print('Classes to predict: ', data.target_names)

data
```

#Ex2: Extracting data attributes (X) and target/ class labels (y)

```python
#To do: type your code here:
#Extracting data atributes
X = data.data
print("Number of exampes in the data:", X.shape[0])
#Extracting target/ class atributes
Y = data.target

X[:4]
```

#Ex3: Create train and test sets (train_test_split), 25% for testing

```python
#To do: type your code here:
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25,random_state = 42)
```

#Ex4: Create a Decision Tree (DT) classifier based on entropy (DecisionTreeClassifier)

```python
#To do: type your code here:
clf = DecisionTreeClassifier(criterion = 'entropy')
```

#Ex5: Train the DT classifier

```python
#To do: type your code here:
clf.fit(X_train, Y_train)
```

#Ex6: Predict labels on the test set.

```python
#To do: type your code here:
Y_pred = clf.predict(X_test)

Y_pred
```
